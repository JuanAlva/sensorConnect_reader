```python
import pandas as pd
#import matplotlib as plt

file_path = 'SensorConnectData.csv'
#csv_file = "SensorConnectData.csv"

def get_header():
    header_flag = 0
    header_table = []
    header = ["time"]
    raw_header_number = 0
    with open(file_path, 'r') as file:
        it = -4
        for line in file:
            it += 1

            #print(line.strip())
            line = line.strip()
            if line == "Channel,Type,SampleRate,Equation,Coefficients,Unit,UnitSymbol,WhereApplied,WhenApplied":
                header_flag = 1
                continue

            if line == "":
                header_flag = 0

            if header_flag == 1:
                print(line)
                header_table.append(line)

            if  line == "DATA_START":
                raw_header_number = it
            
            

    # automatically close when end up block line

    #print(header_table)
    
    for i in header_table:
        #print(i)
        raw = i.split(":")
        if "," in raw[1]:
            raw_clean = raw[1].split(",")
            print(raw_clean[0])
            header.append(raw_clean[0])
        else:
            valid = raw[2].split(",")
            print(raw[1],":",valid[1])
            header.append(valid[1])
    # print(header)
    return header, raw_header_number
    
header_list, raw_header_number = get_header()

print(get_header())


df = pd.read_csv(file_path, sep=',', header=raw_header_number)

print(df)
# with open(file_path, mode='r', newline='', encoding='utf-8') as csvfile:
#     # Create a reader object
#         print(csvfile)
#     # csv_reader = pd.read_csv(csvfile, delimiter=',')
#     # print(csv_reader)
```
