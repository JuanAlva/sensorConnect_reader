```python
import numpy as np

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
escalar = 2

# 1. Multiplicación elemento a elemento
prod_elementos = v1 * v2  # Resultado: [ 4 10 18]

# 2. Producto escalar (dot product) - Resultado un solo número
prod_escalar = np.dot(v1, v2)  # 1*4 + 2*5 + 3*6 = 32
prod_escalar_alternativo = v1 @ v2  # Uso del operador @

# 3. Multiplicación de vector por escalar
prod_vector_escalar = v1 * escalar  # Resultado: [2 4 6]
```
