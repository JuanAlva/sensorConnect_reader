```python
import numpy as np

def static_calibration(acc, gyro, g=9.80665):
    """
    Calibración estática básica de IMU

    acc  : Nx3 acelerómetro [m/s^2]
    gyro : Nx3 giróscopo [rad/s]

    returns:
        acc_bias, gyro_bias
        acc_noise_std, gyro_noise_std
    """

    # === Bias ===
    gyro_bias = np.mean(gyro, axis=0)

    acc_mean = np.mean(acc, axis=0)
    acc_bias = acc_mean - np.array([0, 0, g])

    # === Ruido (desviación estándar) ===
    gyro_noise_std = np.std(gyro - gyro_bias, axis=0)
    acc_noise_std  = np.std(acc - acc_mean, axis=0)

    return acc_bias, gyro_bias, acc_noise_std, gyro_noise_std
```
